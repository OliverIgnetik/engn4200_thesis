```python
# stft with default hop length and window size 

sr = 44100 # 44.1kHz sampling frequency 
hop_length = 512  # overlap of 1536 samples
n_fft = 3004 # default window size 

S = librosa.stft(x,hop_length=hop_length,n_fft = n_fft)
print(S.shape)
```

```python
S_db = librosa.amplitude_to_db(abs(S))
plt.figure(figsize=(14, 4))
_ = librosa.display.specshow(S_db, sr=sr, x_axis='time', y_axis='log')
_ = plt.colorbar()
```

### NMF decomposition

NMF extracts unique events 

```python
# seperate the harmonic STFT into magnitude and phase
X, X_phase = librosa.magphase(S)

# NMF extracts unique events
n_components = 10
# NMF decomposition 
W, H = librosa.decompose.decompose(X, n_components=n_components, sort=True)

print(W.shape)
print(H.shape)
```

```python
plt.figure(figsize=(13, 7))
logW = np.log10(W)
for n in range(n_components):
    plt.subplot(np.ceil(n_components/2.0), 2, n+1)
    plt.plot(logW[:,n])
    plt.ylim(-3, logW.max())
    plt.xlim(0, W.shape[0])
    plt.ylabel('Component %d' % n)
```

## Spectral peak picking - f0 estimation

```python
# extract frequency from component dfts
def freq_from_dft(dft):
    # paramaters for peak picking
    pre_max = 10 
    post_max = 10 
    pre_avg = 10
    post_avg = 10
    delta = 0.5
    wait = 10
    # convert magnitude of dft 
    dft_mag = abs(dft)
    # peak picking algorithm from librosa 
    peak_index = librosa.util.peak_pick(dft_mag, pre_max, post_max, pre_avg, post_avg, delta, wait)
    
    # freq resolution is given by the window length of the original STFT
    return sr*peak_index/n_fft
```

```python
# find the fundamental frequency
print('---------Peaks found from algorithm------------')
f0_freq = []
for n in range(n_components):
    dft = W[:,n]
    freq_est = freq_from_dft(dft)
    f0_freq.append(freq_est[0])
    print('peaks from component', n,':',freq_est)

MIDI_nums = np.around(librosa.hz_to_midi(f0_freq))
print('---------MIDI estimates------------')
MIDI_nums
```

```python
plt.figure(figsize=(13, 7))
for n in range(n_components):
    plt.subplot(np.ceil(n_components/2.0), 2, n+1)
    plt.plot(H[n])
    plt.ylim(0, H.max())
    plt.xlim(0, H.shape[1])
    plt.ylabel('Component %d' % n)
```

```python
# estimating onset times from activation components 
hop_length = 512
initial_time = 15 # seconds
pre_max = 10 
post_max = 10 
pre_avg = 50
post_avg = 50
delta = 1.2
wait = 10

onset_times = []
for H_n in H:
    # peak is picked based on satisfying these conditions: 
    # x[n] == max(x[n - pre_max:n + post_max])
    # x[n] >= mean(x[n - pre_avg:n + post_avg]) + delta
    # n - previous_n > wait
    peak_frame_no = librosa.util.peak_pick(H_n, pre_max, post_max, pre_avg, post_avg, delta, wait)
    # convert from frame number to sample number
    samples_no = peak_frame_no*hop_length
    # convert from samples to time and append to onset times array 
    onset_times.append(float(librosa.samples_to_time(samples_no, sr=sr)+initial_time)) 
onset_times
```

```python
# compare with ground truths 
df = pd.DataFrame(list(zip(onset_times,MIDI_nums)))
df.columns = ground_truths_sample.columns
df
```

# AUTOCORRELATION METHOD FOR PITCH TRANSCRIPTION

```python
# estimate pitch using the autocorrelation method
def estimate_pitch(segment, sr, fmin=50.0, fmax=2000.0):
    # Compute autocorrelation of input segment.
    r = librosa.autocorrelate(segment)
    
    # Define lower and upper limits for the autocorrelation argmax.
    i_min = sr/fmax
    i_max = sr/fmin
    r[:int(i_min)] = 0
    r[int(i_max):] = 0
    
    # Find the location of the maximum autocorrelation.
    i = r.argmax()
    f0 = float(sr)/i
    return f0
```

```python
# recreate the time domain components 
components = []
for n in range(n_components):
    
    # Re-create the STFT of a single NMF component.
    Y = scipy.outer(W[:,n], H[n])*X_phase
    
    # Transform the STFT into the time domain.
    y = librosa.istft(Y)
    components.append(y)
```

```python
# estimate f0 using ACR method
f0_freq = []
count = 0 
for component in components: 
    f0_freq.append(estimate_pitch(component,sr))
    count+=1
    
MIDI_nums = np.around(librosa.hz_to_midi(f0_freq))
```

```python
# print MIDI nums 
print(MIDI_nums)
```

# Detect onsets - with onset_strength function

```python
# detect onsets 
hop_length = 100
onset_env = librosa.onset.onset_strength(x, sr=sr, hop_length=hop_length)
# plot the 1st 1000 samples
plt.plot(onset_env);
plt.xlim(0, len(onset_env));
```

```python
# find the onsets 
hop_length = 200

# peak is picked based on satisfying these conditions: 
# x[n] == max(x[n - pre_max:n + post_max])
# x[n] >= mean(x[n - pre_avg:n + post_avg]) + delta
# n - previous_n > wait

onset_samples = librosa.onset.onset_detect(x,
                                           sr=sr, units='samples', 
                                           hop_length=hop_length, 
                                           backtrack=False,
                                           pre_max=20,
                                           post_max=20,
                                           pre_avg=100,
                                           post_avg=100,
                                           delta=0.02,
                                           wait=0)
# convert to time
onset_times = librosa.samples_to_time(onset_samples, sr=sr)
# we took samples from 15 to 17 seconds
onset_times = onset_times + 15
onset_times
```
