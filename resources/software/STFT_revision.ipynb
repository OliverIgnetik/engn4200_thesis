```python
import scipy, librosa
import librosa.display
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import IPython.display as ipd 
from glob import glob
```

# STFT Parameters

```python
# total number of samples 

audio_files = glob('*.mp3')
x , sr = librosa.load(audio_files[0])

# Convert interval into frame numbers for slicing a part of the song
start = 40 
end = 50
sample_progression = librosa.time_to_samples(np.arange(start,end,1), sr = sr)
#  grab the starting and ending frames
start_sample = sample_progression[0]
end_sample = sample_progression[-1]

sample = x[start_sample:end_sample]
```

```python
# Checking the audio
ipd.Audio(sample, rate=sr)
```

```python
N = len(sample)
print('The total number of samples, N : '+str(N))
print('The sampling frequency, Fs : '+ str(sr)+ ' Hz')
```

## Changing the window size 

### Trade off between temporal and frequency resolution

```python
# hop length is the space in between the windows 
hop_length = 256

# n_fft is the length of the window ie. the number of samples in each DFT
n_fft1 = 30000
print('Window is large: ', n_fft1)
n_fft2 = 1000
print('Window is narrow: ', n_fft2) 

# spectrogram with better frequency resolution
S1  = librosa.feature.melspectrogram(sample , sr = sr, n_fft=n_fft1, hop_length=hop_length)
logS1 = librosa.power_to_db(abs(S1))

# spectrogram with better temporal resolution
S2  = librosa.feature.melspectrogram(sample , sr = sr, n_fft=n_fft2, hop_length=hop_length)
logS2 = librosa.power_to_db(abs(S2))
```

```python
plt.figure(figsize=(15, 5))
librosa.display.specshow(logS1, sr=sr, hop_length=hop_length, x_axis='time', y_axis='mel')
plt.colorbar(format='%+2.0f dB')
plt.title('Better frequency resolution')

frequency_resolution = sr/n_fft1
temporal_resolution = n_fft1*1/sr
print('Frequency resolution: ',round(frequency_resolution,2), 'Hz')
print('Temporal resolution: ',round(temporal_resolution,2), 's')
```

Look at the narrowband effect of each frequency filter
* temporal information lost
* the peak intensity measured is over 30dB
* as there is less temporal granularity the energy is spread in time

```python
plt.figure(figsize=(15, 5))
librosa.display.specshow(logS2, sr=sr, hop_length=hop_length, x_axis='time', y_axis='mel')
plt.colorbar(format='%+2.0f dB')
plt.title('Better Temporal Resolution')

frequency_resolution = sr/n_fft2
temporal_resolution = n_fft2*1/sr
print('Frequency resolution: ',round(frequency_resolution,2), 'Hz')
print('Temporal resolution: ',round(temporal_resolution,2), 's')
```

The maximum frequency intensity is much lower in this case
* note onsets are easier to spot
* There are clear intervals with little to no noise

## Sample chromagram

```python
# Chromagram 
C  = librosa.feature.chroma_cqt(sample , sr = sr)
librosa.display.specshow(C,y_axis='chroma', x_axis= 'time')
plt.colorbar()
plt.title('CQT Chromagram sample')
```

We can see that the pitch determination is quite clear: 
* The key is quite clearly F 
* The musical structures are more apparent
