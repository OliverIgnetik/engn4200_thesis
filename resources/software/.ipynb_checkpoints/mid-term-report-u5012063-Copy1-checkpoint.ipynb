# Exploratory Data Analysis

As in other audio-related applications, the most popular
tool for describing the time-varying energy across different
frequency bands is the short-time Fourier transform (STFT),
which, when visualized as its magnitude, is known as the
spectrogram. It is defined mathematically as: 

$$ X(t,k) = \sum_{n=0}^{N-1}  w[n] \cdot x[n+t+N/2]\cdot e^{-j2 \pi \frac{kn}{N}}  $$  

They are computed over k frequency basis vectors and t time frames determined by the hop length. Usually a hanning window ($w[n]$) is used to smooth out discontinuties.

A Short Time Fourier Transform (STFT) will be used in the NMF approach. However, the time resolution of the spectrogram must be balanced with the frequency resolution. [3]

```python
# we need semitone resolution 
b,c = librosa.note_to_hz(['B3','C4'])
semitone = np.round(c-b,2)
print('semitone resolution:',semitone,'Hz')
print('FFT needs', int(sr/semitone), 'samples')
```

```python
print('time resolution:', int(sr/semitone)/sr,'s')
```

```python
# stft with default hop length and window size 

sr = 44100 # 44.1kHz sampling frequency 
hop_length = 512  # spacing in between dfts 
n_fft = 3004 # window size

S = librosa.stft(x,hop_length=hop_length, n_fft=n_fft)
print(S.shape)
```

```python
# Use of the db scale to replicate the way humans interpret sound 
S_db = librosa.amplitude_to_db(abs(S))
plt.figure(figsize=(14, 4))
_ = librosa.display.specshow(S_db, sr=sr, x_axis='time', y_axis='log')
_ = plt.colorbar()
```

# Non-negative Matrix Factorization (NMF) approach

NMF is an algorithm that factorizes a nonnegative matrix, $X$, into a product of two nonnegative matrices, $W$ and $H$. It is an unsupervised iterative algorithm that minimizes a distance between $X$ and the product $W \times H$:

$$ \min_{W, H} d(X, W \times H) $$

If $X$ has dimensions $M$ by $N$, then $W$ will have dimensions $M$ by $R$, and $H$ will have dimensions $R$ by $N$, where inner dimension $R$ is the *rank* or *number of components* of the decomposition. When applied to a musical signal, the NMF can decompose the signal into separate note events. Therefore, NMF is quite useful and popular for tasks such as transcription and source separation.The input, $X$, is often a magnitude spectrogram. In such a case, the columns of $W$ represent spectra of note events, and the rows of $H$ represent temporal envelopes of the same note events. [4]

```python
# seperate the harmonic STFT into magnitude and phase
# NMF will only factorize the magnitude part of the STFT
X, X_phase = librosa.magphase(S)

# NMF extracts unique events
n_components = number_notes

# NMF decomposition 
W, H = librosa.decompose.decompose(X, 
                                   n_components=n_components,
                                   sort=True)

print('Shape of Dictionary and Activation matrices')
print('-'*60)
print(W.shape)
print(H.shape)
```

```python
# show some components 
plt.figure(figsize=(13, 7))
# take the logarithm of each spectral component
for n in range(4):
    plt.subplot(4, 1, n+1)
    plt.plot(W[:,n])
    plt.xticks(fontsize=12)
    plt.xlim(0, 300)
    plt.ylabel(f'Template {n+1}', fontsize=12)
    plt.yticks(np.arange(0,15,5), fontsize=12)
    
plt.tight_layout()
plt.show()
```

## f0 estimation - dft analysis

Most musical instruments—including string-based instruments such as guitars, violins, and pianos, as well as instruments
based on vibrating air columns such as flutes, clarinets, and
trumpets—are explicitly constructed to allow performers to
produce sounds with easily controlled, locally stable fundamental periods. Such a signal is well described as a harmonic
series of sinusoids at multiples of a fundamental frequency,
and results in the perception of a musical note (a single perceived
event) at a clearly defined pitch in the mind of the listener.
With the exception of unpitched instruments like drums, and
a few inharmonic instruments such as bells, the periodicity of
individual musical notes is rarely ambiguous, and thus equating
the perceived pitch with fundamental frequency is common.
[5]

One way to do this is by computing the difference between peaks in the power spectrum.

```python
# extract frequency from component dfts
def estimate_pitch(power_spectrum,fmin=20,fmax=4300,n_fft=3004,sr=44100):
    # parameters for peak picker   
    # peak is picked based on satisfying these conditions: 
    # x[n] == max(x[n - pre_max:n + post_max])
    # x[n] >= mean(x[n - pre_avg:n + post_avg]) + delta
    # n - previous_n > wait   
    pre_max = 10 
    post_max = 10 
    pre_avg = 10
    post_avg = 10
    delta = 0.5
    wait = 10
    # evaluate maximum peaks for each power spectrum 
    peaks = librosa.util.peak_pick(power_spectrum,
                                   pre_max,
                                   post_max,
                                   pre_avg,
                                   post_avg,
                                   delta,
                                   wait)
    
    # we only need 3 harmonics to calculate the f0 
    peaks = peaks[:5]
    diffs = []
    # find the difference in between each of the peaks 
    for j in range(len(peaks)-1):
        diffs.append(peaks[j+1]-peaks[j])
    
    # find the average of the differences which corresponds to the fundamental frequency 
    i_index = np.floor(np.average(diffs))
    # find the frequency 
    f0 = sr*i_index/n_fft
    return f0
```

```python
# find the fundamental frequency
f0_freq = []
for n in range(n_components):
    power_spectrum = W[:,n]
    freq_est = estimate_pitch(power_spectrum)
    f0_freq.append(freq_est)

MIDI_nums = np.around(librosa.hz_to_midi(f0_freq))
print('MIDI estimates')
print('-'*60)
MIDI_nums
```

## Note onsets 

The notes display characteristic curves (attack, sustain, decay, release) for piano notes as can be seen in the graphs below.

```python
# show some activations for different notes 
plt.figure(figsize=(13, 7))
for n in range(4):
    plt.subplot(4, 1, n+1)
    plt.plot(H[n])
    plt.ylim(0, 200)
    plt.yticks(fontsize=12)
    plt.xticks(fontsize=12)
    plt.xlim(0, H.shape[1])
    plt.ylabel('Activation %d' % (n+1), fontsize=12)
    
plt.tight_layout()
plt.show()
```

```python
# estimating onset times from activation components 
hop_length = 512
attack_frames = 3

# peak picking parameters
# peak is picked based on satisfying these conditions: 
# x[n] == max(x[n - pre_max:n + post_max])
# x[n] >= mean(x[n - pre_avg:n + post_avg]) + delta
# n - previous_n > wait   
pre_max = 10 
post_max = 10 
pre_avg = 50
post_avg = 50
delta = 2
wait = 10

onset_times = []

for H_n in H: 
    peak_frame_no = librosa.util.peak_pick(H_n,
                                           pre_max,
                                           post_max,
                                           pre_avg,
                                           post_avg,
                                           delta,
                                           wait)
    # account for attack phase in note 
    peak_frame_no -= attack_frames
    # convert from frame number to sample number
    samples_no = peak_frame_no*hop_length
    # convert from samples to time and append to onset times array 
    onset_times.append(float(librosa.samples_to_time(samples_no, sr=sr)+initial_time)) 
    
print('Onset times for each note')
print('-'*60)
print(onset_times)
```

```python
# NMF transcription dataframe 
NMF_transcription = pd.DataFrame(list(zip(onset_times,MIDI_nums)))
NMF_transcription.columns = ground_truths_sample.columns
NMF_transcription
```

# Autocorrelation/CEPSTRUM segmentation approach 

## Onset Detection and segmentation

A widely used approach to onset detection in the frequency domain is the spectral flux, where changes of pitch and timbre are detected by analyzing the
signal’s short-time spectrum. [5]

The LibROSA library has an onset detection method which determines onsets of notes by calculating the change in spectral energy and calculating the resulting onset curve shown below.

The onsets can then be used to segment the signal between different note onsets.

```python
# plot the onset envelope that will be used in the onset detect method
# SPECTRAL FLUX is used to find the onset sample numbers
# thus a dft is needed and we need to specify the hop length
hop_length = 512 

# Find the onsets using spectral flux 
onset_env = librosa.onset.onset_strength(x, 
                                         sr=sr,
                                         hop_length=hop_length)

# plot the onset envelope 
plt.plot(onset_env);
plt.ylim(0,20);
plt.xlim((0,200));
```

```python
# find the onsets
# hop length of 512 gives 174 frames
hop_length = 512

# peak is picked based on satisfying these conditions: 
# x[n] == max(x[n - pre_max:n + post_max])
# x[n] >= mean(x[n - pre_avg:n + post_avg]) + delta
# n - previous_n > wait

onset_samples = librosa.onset.onset_detect(x,
                                           sr=sr, 
                                           units='samples', 
                                           hop_length=hop_length, 
                                           pre_max=20,
                                           post_max=20,
                                           pre_avg=50,
                                           post_avg=50,
                                           delta=0.02,
                                           wait=0)

# define onset_boundaries for segmenting audio
onset_boundary_samples = np.concatenate([onset_samples, [N]])

# for mir_eval calculation
onset_times = librosa.samples_to_time(onset_samples, sr=sr)
# add the initial time
onset_times += initial_time

print('Onset times')
print('-'*60)
print(onset_times)
```

## f0 estimation

### cepstrum method
A cepstrum is the result of taking the inverse Fourier transform (IFT) of the logarithm of the estimated spectrum of a signal. It was first defined in a paper by Tukey and Healey in 1963. Cepstrum pitch determination is effective at finding periodics in discrete fourier transforms. It is defined mathematically as : 

$$ power\:cepstrum= |\mathfrak{F}^{-1}\{log(|\mathfrak{F}\{f(t)\}|^{2})\}|^{2}$$  

The independent variable of the cepstral graph is called the **quefrency**. Lower quefrencies correspond to higher frequncies. It is a common industry standard in pitch determination algorithms. [6]

### autocorrelation method
The autocorrelation of a discrete signal is mathematically defined as : 

$$ f(\tau) = \sum_{n=t+1}^{t+W}x[n]x[n+\tau]  $$  

The highest peak in the autocorrelation function corresponds to the fundamental frequency of the sound signal. It is often used in the YIN algorithm as a first step in determining fundamental frequencies in sound signals. [7]  

In librosa a damped autocorrelation function is used which improves the accuracy of the algorithm by ensuring that the first relative maximum will have the highest peak.

```python
from scipy.fftpack import fft, ifft 

def estimate_pitch(segment, sr, f_min=20, f_max=4200, method='auto'):
    
    # autocorrelation method 
    if method == 'auto':
        # Compute autocorrelation of input segment.
        r = librosa.autocorrelate(segment)
        # Define lower and upper limits for the autocorrelation argmax.
        i_min = sr/f_max
        i_max = sr/f_min
        r[:int(i_min)] = 0
        r[int(i_max):] = 0
        # Find the location of the maximum autocorrelation.
        i = r.argmax()
        f0 = float(sr)/i
        return f0 
    
    # cepstrum method     
    elif method == 'cepstrum':
        # Compute power spectrum 
        power_spectrum = np.abs(fft(segment))**2
        # compute cepstrum 
        CEPSTRUM = np.abs(ifft(np.log10(power_spectrum)))**2
        # set indexes
        start = int(sr/f_max)
        end = int(sr/f_min)
        # narrow cepstrum to frequency range of interest 
        CEPSTRUM = CEPSTRUM[start:end]
        peak_ix = CEPSTRUM.argmax()
        # find the fundamental frequency 
        f0 = sr/(start + peak_ix)
        return f0
```

```python
# sine tone for transcription playback
def generate_sine(f0, sr, n_duration):
    # set a duration time for the tone
    n = np.arange(n_duration)
    return 0.2*np.sin(2*np.pi*f0*n/float(sr))
```

```python
# calculate the pitch estimations for the two different methods
MIDI_nums_auto = []
MIDI_nums_cepstrum =[]    
playback = []

for i in range(len(onset_boundary_samples)-1):
        # slice audio segments 
        n0 = onset_boundary_samples[i]
        n1 = onset_boundary_samples[i+1]
        segment_duration = n1 - n0
    
        # estimate pitch for segment using autocorrelation
        f0 = estimate_pitch(x[n0:n1], sr,method='auto')
        MIDI_nums_auto.append(int(librosa.hz_to_midi(f0)))
        playback.append(generate_sine(f0,sr,segment_duration))
        
        
        # estimate pitch for segment using cepstrum         
        f0 = estimate_pitch(x[n0:n1], sr,method='cepstrum')
        MIDI_nums_cepstrum.append(int(librosa.hz_to_midi(f0)))

# concatenate sine tones for playback format
playback = np.concatenate(playback)
```

```python
# store autocorrelation results in a dataframe
auto_transcription = pd.DataFrame(list(zip(onset_times,MIDI_nums_auto)))
auto_transcription.columns = ground_truths_sample.columns
auto_transcription
```

```python
# store CEPSTRUM results in a dataframe
CEPSTRUM_transcription = pd.DataFrame(list(zip(onset_times,MIDI_nums_cepstrum)))
CEPSTRUM_transcription.columns = ground_truths_sample.columns
CEPSTRUM_transcription
```

```python
# play back the audio for the autocorrelation method
ipd.Audio(playback, rate=sr)
```

# Evaluation of different methods

mir_eval is a python package that can be used to evaluate the accuracy of transcription algorithms with metrics other then accuracy which is often not correlated with human experience. [9] A link to the repository can be found here: http://craffel.github.io/mir_eval/

```python
# mir eval is used to evaluate accuracy and other standardized metrics 
# that are used by MIREX
import mir_eval 
```

```python
# place values in appropriate format 
ref_time = ground_truths_sample.OnsetTime.values
ref_freqs = librosa.midi_to_hz(ground_truths_sample.MidiPitch.values)

# NMF values
NMF_time = NMF_transcription.OnsetTime.values
NMF_freqs = librosa.midi_to_hz(NMF_transcription.MidiPitch.values)

# auto_correlation values 
auto_time = auto_transcription.OnsetTime.values
auto_freqs = librosa.midi_to_hz(auto_transcription.MidiPitch.values)

# cepstrum values 
cepstrum_time = CEPSTRUM_transcription.OnsetTime.values
cepstrum_freqs = librosa.midi_to_hz(CEPSTRUM_transcription.MidiPitch.values)

# evaluate the quality of the transcription 
print('NMF approach accuracy metrics')
print('-'*60)
print(mir_eval.melody.evaluate(ref_time,ref_freqs,NMF_time,NMF_freqs))
print('-'*60)
print('Autocorrelation approach accuracy metrics')
print('-'*60)
print(mir_eval.melody.evaluate(ref_time,ref_freqs,auto_time,auto_freqs))
print('-'*60)
print('CEPSTRUM approach accuracy metrics')
print('-'*60)
print(mir_eval.melody.evaluate(ref_time,ref_freqs,cepstrum_time,cepstrum_freqs))
```

## Remarks 

We can see that the autocorrelation method performs best overall but is only applicable to single note melodies. The NMF approach however can be extended to polyphonic cases, which is why this approach will be pursued in the future.

# Future Work 

The outcomes of this report are only applicable to monophonic music. Ultimately the goal of this research project is to explore polpyhonic music. There are two promising avenues that will be pursued in the future : 

* NMF approaches 
* Neural Network (NN) approaches 

An preliminary attempt at using NN [10] to solve polyphonic music can be found here : 

https://gist.github.com/OliverIgnetik/f502393b6049cdd4fee358baf3cc89c7 [*HTML jupyter notebook format*]

However, in order to improve the **NN** model, future work is needed: 

* Solve the class imbalance / sparsity problem. There are many inactive notes in every frame. A more appropriate evaluation metric is needed such as weighted class accuracy.
* Long Short-Term Models for NN to learn temporal dependencies in the model.
* Investigate using multiple spectrograms as input to the model.

The **NMF** approach may be improved by exploring : 

* Hidden Markov Models (HMMs) to model temporal dependencies between note transitions 
* Investigating source separation literature to help with disentangling energy from multiple pitches

Finally some work has been done to characterize chords or groups of notes by features [10] such as spectral entropy and chroma characeteristics which can be found here : 

https://gist.github.com/OliverIgnetik/91549a6d75727ff932138e2f6a1e1aee [*HTML jupyter notebook format*]

The aim is that by characterizing groups of notes by certain characteristics the total number of possible choices can be reduced and may be incorporated in designing the observation matrix for a HMM model. [11]
