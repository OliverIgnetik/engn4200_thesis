# Introduction 

This notebook includes the code for transcribing a music sample from the MAPS database using Non-negative Matrix Factorization. 
This sample recording is a chromatic scale played on a Bechstein D 280 in a concert hall recording environment retrieved from the MAPS database. [1]

Ground truths are recorded using the disklavier technology which records pitch, onset and offset times of each note played. 

**MAPS database url**: http://www.tsi.telecom-paristech.fr/aao/en/category/database/ 

## LibROSA package 

**LibROSA github page** : https://librosa.github.io/librosa/ 

LibROSA is a python package for music and audio analysis. It provides the building blocks necessary to create music information retrieval systems. It has many useful features such as peak picking algorithms, factorization methods, spectrogram computation and beat onset calculation. [2]

# Hyperparameter Tuning

## Class based approach

```python
class NMF_model:
    # static variables
    audio = audio
    # variables to extract from audio recording      
    number_notes = 10 
    initial_time = 15
    # sampling rate     
    sr = 44100 
    
    def __init__(self, hop_length=512, n_fft=3004):
        # set parameters to investigate 
        self.hop_length = hop_length
        self.n_fft = n_fft
        
    def make_stft(self):
        self.spectrogram = librosa.stft(audio, hop_length=self.hop_length, n_fft=self.n_fft)
        return 
    
    def NMF_decomposition(self):
        # NMF will only factorize the magnitude part of the STFT
        X, X_phase = librosa.magphase(self.spectrogram)
        # NMF decomposition 
        W, H = librosa.decompose.decompose(X, n_components=number_notes, sort=True)
        self.dictionary = W 
        self.activations = H
        return 
        
    def estimate_onsets(self,onsets_max=10, onsets_avg=50, onsets_delta=2, onsets_wait=10, attack_frames=3):
        try: 
            # peak picking parameters for onset time 
            pre_max, post_max = tuple([onsets_max]*2) 
            pre_avg, post_avg = tuple([onsets_avg]*2)
            delta = onsets_delta
            wait = onsets_wait
            
            # container for onset times
            onsets = []

            for activation in self.activations: 
                peak_frame_no = librosa.util.peak_pick(activation, pre_max, post_max, pre_avg,
                                                       post_avg, delta, wait)
                # account for attack phase in note 
                peak_frame_no -= attack_frames
                # convert from frame number to sample number
                samples_no = peak_frame_no*self.hop_length
                # onset time 
                onset_time = float(librosa.samples_to_time(samples_no, sr=sr)+initial_time)
                # convert from samples to time and append to onset times array 
                onsets.append(onset_time) 

            self.onsets = onsets
        except:
            self.onsets = None
    
    def estimate_pitches(self, pitches_max=10, pitches_avg=10, pitches_delta=0.5, pitches_wait=10, harmonics=3):
        try:
            # find the fundamental frequency
            pitches = []
            for n in range(number_notes):
                # perform peak picking algorithm for each spectral template
                power_spectrum = self.dictionary[:,n]
                
                # parameters for pitch peak picker   
                pre_max, post_max = tuple([pitches_max]*2)
                pre_avg, post_avg = tuple([pitches_avg]*2)
                delta = pitches_delta
                wait = pitches_wait
                
                # evaluate maximum peaks for each power spectrum 
                peaks = librosa.util.peak_pick(power_spectrum, pre_max, post_max, 
                                               pre_avg, post_avg, delta, wait)
                
                # slice subset of peaks
                # technically we only need 3 or more                 
                peaks = peaks[:harmonics]
                diffs = []
                # find the difference in between each of the peaks 
                for j in range(len(peaks)-1):
                    diffs.append(peaks[j+1]-peaks[j])

                # find the average of the differences which corresponds to the fundamental frequency 
                i_index = np.floor(np.average(diffs))
                # find the frequency 
                pitch = sr*i_index/self.n_fft
                pitches.append(pitch)

            self.pitches = pitches
        except: 
            self.pitches = None
```

```python
# function for evaluating accuracies 
def evaluate_accuracy(ref_time, ref_freqs, model):
    try:
        accuracy_metrics=mir_eval.melody.evaluate(ref_time,ref_freqs, model.onsets, model.pitches)
        return round(accuracy_metrics['Overall Accuracy'],3)
    except:
        print('Parameter value incompatible')
```

```python
# PARAMETER TUNING
window_sizes = {
    'name': 'window size',
    'data':[3004, 9012, 12016]
}

# attack frames models note evolution
attack_frames= {
    'name' : 'attack frames',
    'data':[0,7,15]
}

# threshold function for peak picking
pitches_delta = {
    'name': 'threshold function pitches',
    'data':[0.5,0.7,0.9]
}

# number of harmonics used in calculation of f0 
harmonics = {
    'name': 'harmonics',
    'data':[5,7,9]
}

params_list = [window_sizes, hop_lengths, attack_frames, pitches_delta, harmonics]
```

```python
# main script
# list to store values 
window_sizes_vals = []
window_sizes_accs = []

attack_frames_vals = []
attack_frames_accs = []

pitches_delta_vals = []
pitches_delta_accs = []

harmonics_vals = []
harmonics_accs = []

# reference values for use in evaluate function  
ref_time = ground_truths_sample.OnsetTime.values
ref_freqs = librosa.midi_to_hz(ground_truths_sample.MidiPitch.values)

for param in params_list:
    if param['name']=='window size':
        for val in param['data']:
            window_sizes_vals.append(val)
            model = NMF_model(n_fft=val)
            # construct the stft and NMF decomposition      
            model.make_stft()
            model.NMF_decomposition()     
            # pitches     
            model.estimate_pitches()
            # onsets     
            model.estimate_onsets()
            accuracy = evaluate_accuracy(ref_time, ref_freqs, model)
            window_sizes_accs.append(accuracy)
            
    elif param['name']=='attack frames':
        for val in param['data']:
            attack_frames_vals.append(int(val))
            model = NMF_model()
            # construct the stft and NMF decomposition      
            model.make_stft()
            model.NMF_decomposition()     
            # pitches     
            model.estimate_pitches()
            # onsets     
            model.estimate_onsets(attack_frames=val)
            accuracy = evaluate_accuracy(ref_time, ref_freqs, model)
            attack_frames_accs.append(accuracy)
            
    elif param['name']=='threshold function pitches':
        for val in param['data']:
            pitches_delta_vals.append(val)
            model = NMF_model()
            # construct the stft and NMF decomposition      
            model.make_stft()
            model.NMF_decomposition()     
            # pitches     
            model.estimate_pitches(pitches_delta=val)
            # onsets     
            model.estimate_onsets()
            accuracy = evaluate_accuracy(ref_time, ref_freqs, model)
            pitches_delta_accs.append(accuracy)
            
    elif param['name']=='harmonics':
        for val in param['data']:
            harmonics_vals.append(val)
            model = NMF_model()
            # construct the stft and NMF decomposition      
            model.make_stft()
            model.NMF_decomposition()     
            # pitches     
            model.estimate_pitches(harmonics=val)
            # onsets     
            model.estimate_onsets()
            accuracy = evaluate_accuracy(ref_time, ref_freqs, model)
            harmonics_accs.append(accuracy)
```

```python
# plot the information
# indexed from 1
plt.figure(figsize=(10,10))

# window size plot
plt.subplot(221)

plt.plot(window_sizes_vals, window_sizes_accs,marker='o', markerfacecolor='red', markersize=12, color='salmon', linewidth=2)
plt.ylabel('Overall Accuracy',fontsize=16)
plt.xlabel('Window size\n(a)',fontsize=16)
plt.axis([0,15000,0.6,1.05])
plt.xticks(np.arange(2000,18000,4000), fontsize=14)
plt.yticks(fontsize=14)


# pitch threshold function
plt.subplot(222)

plt.plot(pitches_delta_vals, pitches_delta_accs,marker='o', markerfacecolor='darkorange', markersize=12, color='bisque', linewidth=2)
plt.ylabel('Overall Accuracy', fontsize=16)
plt.xlabel('Threshold value\n(b)',fontsize=16)
plt.axis([0.4,1,0,1.1])
plt.xticks(fontsize=14)
plt.yticks(np.arange(0,1.2,0.2),fontsize=14)

# Make the bottom left subplot active in the current 2x2 subplot grid
plt.subplot(223)

plt.plot(harmonics_vals, harmonics_accs, marker='o', markerfacecolor='limegreen', markersize=12, color='darkgreen', linewidth=2)
plt.ylabel('Overall Accuracy',fontsize=16)
plt.xlabel('Number of harmonics \n (c)',fontsize=16)
plt.axis([4,10,0,1.1])
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

# Make the bottom right subplot active in the current 2x2 subplot grid
plt.subplot(224)

# Plot in yellow the % of degrees awarded to women in Education
plt.plot(attack_frames_vals, attack_frames_accs,marker='o', markerfacecolor='dodgerblue', markersize=12, color='powderblue', linewidth=2)
plt.ylabel('Overall Accuracy',fontsize=16)
plt.xlabel('Number of attack frames \n (d)',fontsize=16)
plt.axis([-2,18,0.75,1.04])
plt.xticks(np.arange(0,18,2),fontsize=14)
plt.yticks(np.arange(0.7,1,0.1),fontsize=14)

# Improve the spacing between subplots and display them
plt.tight_layout()
plt.savefig('NMF_tuning.png', dpi=300, bbox_inches='tight')
plt.show()
```
